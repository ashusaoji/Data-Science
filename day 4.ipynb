### Seperate the input and output

```python
x=df[['disp','hp','wt']]
y=df['mpg']
```

```python
x
```

### import the class

```python
from sklearn.linear_model import LinearRegression
```

### create th object

```python
reg=LinearRegression()
```

### train the model

```python
reg.fit(x,y)
```

#### seperate input and output variable

```python
x=df[['Age','EstimatedSalary']]
y=df['Purchased']
```

```python
x
```

```python
y
```

```python
y.value_counts()
```

#### visualize the datam

```python
plt.title('Purchase Data')
plt.xlabel('Age')
plt.ylabel('EstimatedSalary')
plt.grid()
plt.scatter(x['Age'],x['EstimatedSalary'], c=y)
```

#### import the class

```python
from sklearn.tree import DecisionTreeClassifier
```

#### create the object

```python
classifier=DecisionTreeClassifier(random_state=0)
```

#### train the model

```python
classifier.fit(x,y)
```

#### plot the tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,15))
plot_tree(classifier,fontsize=7,feature_names=['age','sal'],class_names=['y','n'],filled=True,rounded=True);
```

#### Predict on new data

```python
new=pd.DataFrame({
    'Age':[23,45,54,67],
    'EstimatedSalary':[45000,89000,23000,134000]
})
```

```python
new
```

```python
classifier.predict(new)
```
